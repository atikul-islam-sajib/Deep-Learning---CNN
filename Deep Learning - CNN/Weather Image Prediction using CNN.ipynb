```python
####### To Extract the image to the data #########
directory = '/content/dataset'
#now we want to extract the cat and dog folder
data, categories = [], ['dew', 'fogsmog', 'frost', 'glaze', 'hail', 'lightning', 'rain', 'rainbow', 'sandstorm', 'snow']

for i in categories:
    
    folder_path = os.path.join(directory,i)
    
    for image in os.listdir(folder_path):
        label = categories.index(i)
        filepath_ = os.path.join(folder_path, image)
        img_arr_  = cv2.imread(filepath_)
        try:
          img_arr_  = cv2.resize(img_arr_, (100, 100))
          ####### append this image vector to the data_ list #######
          data.append([img_arr_, label])
        except:
          print("Error")
          break
    
    print(f"{i} is done.")

random.shuffle(data)
```

```python


####### Split the dataset into X and y ########
X = []
y = []
for value_ in data:
  X.append(value_[0])
  y.append(value_[1])

print("appending is done.")
```

```python
####### Normalized the X and y ########
X = np.array(X)
y = np.array(y)
####### Divided the data by 255 #######
X = X/255
```

```python
####### Train and test split ########
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
####### Create a model #######
model = Sequential()

####### Create first Convolution layer with filters 128 and MaxPooling 2, 2 #######
model.add(Conv2D(filters = 64, kernel_size = (3, 3), activation = 'relu', kernel_initializer = "he_normal", input_shape = X.shape[1:]))
model.add(MaxPool2D(pool_size = (2, 2)))
model.add(Dropout(0.6))

####### Create second Convolution layer with filters 64 and MaxPooling 2, 2 ######
model.add(Conv2D(filters = 32, kernel_size = (3, 3), activation = 'relu', kernel_initializer = "he_normal"))
model.add(MaxPool2D(pool_size = (2, 2)))
model.add(Dropout(0.6))

####### Faltten the model ########
model.add(Flatten())

####### Create first hidden layer with neurons 128 ########
model.add(Dense(units = 200, activation = 'relu', kernel_initializer = HeNormal()))
model.add(Dropout(0.6))

####### Create first hidden layer with neurons 128 ########
model.add(Dense(units = 200, activation = 'relu', kernel_initializer = HeNormal()))
model.add(Dropout(0.5))

####### Create an output layer ########
model.add(Dense(units = 11, activation = 'softmax'))


####### Compile the model #######
model.compile(optimizer = 'Adam', loss = SparseCategoricalCrossentropy(), metrics = ['accuracy'])

###### fit the model and run #######
model.fit(x = X_train, y = y_train, validation_data = (X_test, y_test), epochs = 50, batch_size = 64, verbose = 1)
```
